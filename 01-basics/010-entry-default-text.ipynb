```python
from tkinter import *

root = Tk()

entry = Entry(root, width = 50)
entry.pack()
entry.insert(0, "Enter name here")

def buttonClick():
    myLabel = Label(root, text = "Hello " + entry.get())
    myLabel.pack()

myButton = Button(root, text = "Click here", command = buttonClick, fg = "blue", bg = "#000000")
myButton.pack()

root.mainloop()
```
